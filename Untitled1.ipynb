```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from textblob import TextBlob

plt.style.use("seaborn-v0_8")
```

```python
df = pd.read_csv("student_feedback.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
df.mean().sort_values().plot(kind='bar', color='steelblue')
plt.title("Average Feedback Scores")
plt.ylabel("Average Rating")
plt.show()
```

```python
def rating_sentiment(score):
    if score >= 7:
        return "Positive"
    elif score >= 4:
        return "Neutral"
    else:
        return "Negative"
```

```python
df["sentiment"] = df["course_recommendation_based_on_relevance"].apply(rating_sentiment)
```

```python
df[["course_recommendation_based_on_relevance", "sentiment"]].head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="sentiment", data=df, palette="Set2")
plt.title("Sentiment Distribution of Student Feedback")
plt.show()
```

### Key Insights
- Majority of students provided Positive feedback
- Neutral responses indicate scope for improvement
- Negative feedback highlights areas needing attention
